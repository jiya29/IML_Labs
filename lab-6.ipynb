> # Lab 6_1>>To convert a NumPy array to a Pandas series


```python
import numpy as np
import pandas as pd

a=np.array([20,40,60,80,100])
print("The native numpy array>>\n",a)

b=pd.Series(a)
print("The Pandas Series converted from the numpy array>>\n",b)
```

> # Lab 6_3 >>To create the mean and standard deviation of the data of a given Series


```python
import pandas as pd 
a=pd.Series(data=[10,30,60,90,120,150])
print("The Native Array >> \n",a)

print("The Mean of the array >>\n",(a.mean()))

print("The Standard of the array >>\n",(a.std()))


```

> # Lab 6_4 >>To sort a given Series

```python
import pandas as pd 

a=pd.Series(data=[1000,5000,3000,2000,4000])
print("The Native Array >>\n",a)

b=pd.Series(a).sort_values()
print("The Sorted Series>>\n",b)
```
