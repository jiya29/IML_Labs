> # Lab 3_1 >>to split an array of 14 elements into 3 arrays, each with 2, 4, and 8 elements in the original order using numpy

```python
import numpy as np 
a=np.array([100,110,120,130,140,150])

print("The original array >>\n",a)
print("\n\n")
b=np.split(a,[2,6])

print("The Split an array of 14 elements >>\n",b)
```

> # Lab 3_2 >> To stack arrays horizontally (column wise)

```python
import numpy as np 

a=np.array([100,200,300,400])
b=np.array([500,600,700,800])

print("the original array >>\n",a,b)
print("\n\n")

c=np.hstack((a,b))

print("The Stacking of array horizontally >>\n ",c)
```
