> # Lab 11 >>Write a Pandas program to filter all columns where all entries present, 
check which rows and columns has a NaN and finally drop rows with any NaNs from the given dataset.

```python
#Jiya Chudasama
#216270307021


import pandas as pd
a = pd.read_csv('/kaggle/input/titanic-dataset/Titanic-Dataset.csv')
print(">>The data of passengers :")
print(a.head())
print("\n>> Find  all columns which all entries present:")
print(a.loc[:, a.notnull().all()])
print("\n>>Rows and columns has a NaN:")
print(a.loc[:,a.isnull().any()])
print("\nDrop rows with any NaNs:")
print(a.dropna(how='any'))  
```
