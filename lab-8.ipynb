> # Lab 8_1 >>Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.



```python
#Jiya Chudasama
#216270307021


import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/kaggle/input/stock-market-data-nifty-50-2000-2022/Nifty_50_with_indicators_.csv')
start_date = pd.to_datetime('2021-4-29')
end_date = pd.to_datetime('2021-09-30')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df2 = df.loc[new_df]
plt.figure(figsize=(10,10))
df2.plot(x='Date', y=['open', 'close']);
plt.suptitle('Opening/Closing stock prices of Alphabet Inc.,\n 29-04-2021 to 30-09-2021', fontsize=12, color='blue')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("$ price", fontsize=12, color='black')
plt.show()
```
