> # Lab 5_1 >>To find the maximum and minimum value of a given flattened array


```python
import numpy as np

a=np.arange(5,15)
print("The native array >>\n",a)
print("\n")
print("The Maximum from the array >>\n",(np.amax(a)))
print("\n\n")
print("The Maximum from the array >>\n",(np.amin(a)))
```

> # Lab 5_2 >>To compute the mean, standard deviation, and variance of a given array along the second axis


```python
import numpy as np 
a=np.arange(15)
print("The native array >>\n",a)
print("\n")

a1=np.mean(a)
a2=np.average(a)

print("The mean of the array >>\n ",a1)
print("\n")
a1=np.std(a)
a2=np.sqrt(np.mean((a - np.mean(a)) ** 2 ))

print("The standard of the array >>\n ",a1)
print("\n")
a1=np.var(a)
a2=np.mean((a - np.mean(a)) ** 2 )

print("The Variance of the array >>\n",a1)
```
