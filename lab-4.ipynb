> # Lab 4_1 >> To add, subtract, multiply, divide arguments element-wise


```python
import numpy as np 

print("Addition of two integer number \n\n >>",np.add(10,20))


print("Subtraction of two integer number \n\n >>",np.subtract(40,30))

print("Multiplication of two integer number \n\n >>",np.multiply(50,60))

print("Division of two integer number \n\n >>",np.divide(70,10))
```

> # Lab 4_2 >> To round elements of the array to the nearest integer


```python
import numpy as np

a=np.array([10.8,20.7,30.6,40.5,50.4])

print("The Native  Array >>\n\n ",a)
print("\n\n")
b=np.rint(a)

print("The Round elements of the array to the nearest integer >>\n\n ",b)
```

> # Lab  4_3 >> To calculate mean across dimension, in a 2D numpy array

```python


import numpy as np

# creating 2d array
a = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])


r_mean = np.mean(a, axis=1)

r1_mean = r_mean[0]
print("The Mean of Row 1 >>\n", r1_mean)

r2_mean = r_mean[1]
print(" The Mean of Row 2 >>\n", r2_mean)

r3_mean = r_mean[2]
print(" The Mean of Row 3 >>\n", r3_mean)


col_mean = np.mean(a, axis=0)

col1_mean = col_mean[0]
print(" The Mean of column 1 >> \n", col1_mean)

col2_mean = col_mean[1]
print(" The Mean of column 2 >> \n", col2_mean)

col3_mean = col_mean[2]
print("The Mean of column 3 >> \n", col3_mean)
```

> # Lab 4_4 >>To calculate the difference between neighboring elements, elementwise of a given array


```python
import numpy as np

a=np.array([20,40,30,60,40,80])

print("The Native Array >>\n",a)
print("\n\n")


b=np.diff(a)
print("the difference between neighboring elements, elementwise of a given array >> \n\n",b)
```
